### To make a categorical dataframe

## Its better to drop the column which have more than 40 % of missing value

```python
dfcat.drop('Risk_Segment_During_Rx',inplace=True,axis=1) # 44% data is missing
```

```python
dfcat.drop('Tscore_Bucket_During_Rx',inplace=True,axis=1)# 44% data is missing
```

```python
dfcat.drop('Change_T_Score',inplace=True,axis=1)# 44% data is missing
```

```python
dfcat.drop('Change_Risk_Segment',inplace=True,axis=1)# 65% data is missing
```

```python
dfcat.head()
```

## for the rest of the missing values i will replace it with most frequent occurring value

```python
dfcat['Race'].value_counts()
```

```python
dfcat['Race'].fillna('Caucasian',inplace=True)
```

```python
dfcat['Ethnicity'].value_counts()
```

```python
dfcat['Ethnicity'].fillna('Not Hispanic ',inplace=True)
```

```python
dfcat['Region'].value_counts()
```

```python
dfcat['Region'].fillna('Midwest',inplace=True)
```

```python
dfcat['Ntm_Speciality'].value_counts()
```

```python
dfcat['Ntm_Speciality'].fillna('GENERAL PRACTITIONER ',inplace=True)
```
